```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df=pd.read_csv('hiring.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['experience'].fillna('zero', inplace= True)
df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median(), inplace=True)
```

```python
df
```

```python
from word2number import w2n
df['experience']=df['experience'].apply(w2n.word_to_num)
df
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
test_data = pd.DataFrame({
    "experience": [1, 4, 6, 8, 12],
    "test_score(out of 10)": [7.0, 6.0, 9.0, 8.0, 9.0],
    "interview_score(out of 10)": [8, 9, 8, 7, 9]
})
test_data
```

```python
test_data['predicted_salary']=reg.predict(test_data[['experience','test_score(out of 10)','interview_score(out of 10)']])
test_data
```

```python
test_data.to_csv('predicted_hiring_salary.csv', index=False)
```
